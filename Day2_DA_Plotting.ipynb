```python
#there are 5 sub pyschology nanotec ,medicine ,oncology,antropology
#two students have applied to do masters in any of these five fields with the line plot show in which field they can excel the cutoff of for each sub is 70 (Now students earned each subject 67,45,73,71,57)(43,67,68,53,85)
#s1=67,45,73,71,57
#s2=43,67,68,
sub=["psy","nano","med","onc","anthro"]
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cut=[70,70,70,70,70]
#student 1
plt.scatter(sub,s1,label="student1",color = "k",marker ="p")
plt.scatter(sub,s2,label="student2",color = "b",marker ="H")
plt.plot(sub,cut,label="cutoff",color ="r",linestyle=":")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()
```

```python
#an india vs aus match is on live india has given 289 as target for aus the runs every 5 overs is given as follows india score 25,51,84,131,160,189,220,250,267,289 aus have completed 25 overs score 15,41,94,110,151
#plot overs vs runs for both countries with india has a blue and aus as yellow line give required labels and legend .the marker for india is flag and marker for aus is plus(+).take overs from a range function create arrays for every runs
import numpy as np
india=np.array([25,51,84,131,160,189,220,250,267,289])
aus=np.array([15,41,94,110,151])
overs=np.arange(5,51,5)
plt.plot(overs,india,label="INDIA",marker=">",color="b",linewidth=3)
plt.plot(overs[:5],aus,label="AUSTRALIA",color="y",marker="+",linewidth=3)
plt.xlabel("AUSTRALIA")
plt.ylabel("INDIA")
plt.legend()
plt.title("INDIA VS AUSTRALIA ")
```
